```python
class car:
    def __init__(self,make,modal):
        self.make=make
        self.modal=modal
        self.spped=0
    def start(self):
        print(f"{self.make} {self.modal} started")
        
    def acc(self,mph):
        self.spped+=mph
        print(f"{self.make} {self.modal}  Accelater is {self.spped} mph")
    def brake(self):
        if self.spped>60:
            print(f"{self.make} {self.modal} Apllying Braak")
        
```

```python
# Create object
my_car=car("Tvs","Bajaj")
```

```python
my_car.start() # Inheritence its method car
```

```python
my_car.acc(523)
```

```python
my_car.brake()
```
