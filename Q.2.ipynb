```python
"""
Q.2 Differentiate between Abstraction and Encapsulation. Explain with an example

Ans. Abstraction :Abstraction is a programing example like tv remote we don't need to how to work 
     and how to send single only we need to button power channel in the same way abstraaction hide the complicated 
     stuff and lets only focus on important part
     
     Encapsulation:  encapsulation is like putting your important data (like variables or functions) inside a locked box.
     You can't directly change or see them from anywhere in your code. Instead, you create methods (like keys for the box)
     to access and modify that data. 
```

```python
class car:
    def __init__(self,make,modal):
        self.make=make #Encapsulated attribute
        self.modal=modal # Encapsulated attribute
        self.spped=0
    def start(self):
        print(f"{self.make} {self.modal} started")
        
    def acc(self,mph):# Abstraction mph
        self.spped+=mph
        print(f"{self.make} {self.modal}  Accelater is {self.spped} mph")
    def brake(self):
        if self.spped>60:
            print(f"{self.make} {self.modal} Apllying Braak")
        
```

```python
# creat object
my_car=car("Tvs","Bajaj")
my_car.acc(90)

my_car.brake()
```
